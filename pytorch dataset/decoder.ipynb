```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM


class TextGenerator:
    def __init__(self, model_name="facebook/bart-large-cnn"):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.engine = AutoModelForSeq2SeqLM.from_pretrained(model_name)

        class Config:
            max_length = 100

        self.config = Config()


    def generate(self, **kwargs):
        input_ids, input_masks = [
            kwargs[w] for w in "input_ids, input_masks".strip().split(", ")
        ]

        output = self.engine.generate(
            input_ids=input_ids,
            attention_mask=input_masks,
            max_length=self.config.max_length
        )

        dec = [
            self.tokenizer.decode(ids, skip_special_tokens=False)
            for ids in output
        ]

        output = [
            context.replace("<pad>", "")
                   .replace("</s>", "")
                   .strip()
            for context in dec
        ]

        return output
```

```python
generator = TextGenerator()


text = """
Artificial intelligence is changing the world by enabling computers
to learn from data and make decisions.
"""


# Convert text into tokens
tokens = generator.tokenizer(
    text,
    return_tensors="pt",
    padding=True,
    truncation=True
)
```

```python

# Generate output
result = generator.generate(
    input_ids=tokens["input_ids"],
    input_masks=tokens["attention_mask"]
)
```
